```python
import pandas as pd
import matplotlib.pyplot as plt
from math import log


from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 30, 24
```

```python
'''payment must exceeds the accrued interest each month
accrued interest = PV * (1+i) '''

def how_long(PV, APR, PMT):
    i = (APR/12)/100
    accrued_interest = PV * i
    try:
        N = round(-(log(1-((1000*i)/PMT)))/log(1+i))
        return N
    
    except:
        print(f"Your payment must be greater than accrued_interest which is {'%.2f' % accrued_interest}.")
    
```

```python
how_long(3000, 20,50 )
```

```python
import numpy as np
data = {'PMT':np.arange(40, 1000, 0.3)} 


df = pd.DataFrame(data)
```

```python
df['how_long'] = df.apply(lambda row: how_long(1000, 20, row['PMT']), axis=1)
```
